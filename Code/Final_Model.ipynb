```python
#let's train

df_grand = pd.read_csv('/content/drive/My Drive/StarWords/df_grand_refined.csv', index_col=0)
#removing table rows
df_grand = df_grand[df_grand['label'] != 2]
df_grand['process'] = df_grand.apply(lambda row : process_sent(row['Sentence']), axis = 1 )
print(df_grand.head())
```

```python
# index added
df_grand['index_new'] = range(df_grand.shape[0]) 
df_grand = df_grand.set_index('index_new')

df_grand = df_grand[df_grand['label']!='']

df_grand['label'] = df_grand['label'].astype(np.float64)
df_grand.head()
```

```python
type(df_grand['label'].iloc[0])
```

```python
X = df_grand['process']
y = df_grand['label'].values
idx = range(X.shape[0])

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier
from sklearn.feature_extraction.text import TfidfTransformer


from sklearn.model_selection import train_test_split



X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(X, y, idx, test_size = 0.30, random_state = 3)

model = Pipeline([('vect', CountVectorizer(max_features=150)),
                  ('tfidf', TfidfTransformer()),
                  ('clf', SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3, random_state=42, max_iter=100, tol=None)),
                  ])
print(model)
model.fit(X_train, y_train)



from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 

y_pred = model.predict(X_test)
print('accuracy %s' % accuracy_score(y_pred, y_test))
print(classification_report(y_test, y_pred))

results = confusion_matrix(y_test, y_pred) 
  
print ('Confusion Matrix :')
print(results) 

import pickle
#Saving model 
model_pkl = '/content/drive/My Drive/StarWords/model.pickle'
pickle.dump(model, open(model_pkl, "wb"))
```

```python
y_pred = model.predict(X_train)
print('accuracy %s' % accuracy_score(y_pred, y_train))
print(classification_report(y_train, y_pred))

results = confusion_matrix(y_train, y_pred) 
  
print ('Confusion Matrix :')
print(results) 
```
